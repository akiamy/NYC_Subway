```python
import pandas as pd
import glob
import os

# Set folder path
folder_path = "C:/Users/kylea/iCloudDrive/EAE/Data_Science/final_project/ny_metro_manhattan/raw_data"

# Get all CSV file paths in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Read and concatenate all CSVs into a single DataFrame
combined_df = pd.concat(
    (pd.read_csv(f) for f in csv_files),
    ignore_index=True
)

#quick preview
print(f"Loaded {len(csv_files)} CSVs into a DataFrame with {combined_df.shape[0]:,} rows and {combined_df.shape[1]} columns.")
combined_df.head()
```

```python
# 1. Filter the 'transit_mode' column to only 'subway'
combined_df = combined_df[combined_df["transit_mode"] == "subway"]

# 2. Remove the 'borough' and 'transit_mode' columns
combined_df = combined_df.drop(columns=["borough", "transit_mode"])

# 3. Convert 'transit_timestamp' to datetime
combined_df["transit_timestamp"] = pd.to_datetime(combined_df["transit_timestamp"])

# 4. Extract year, month, day, and hour from 'transit_timestamp'
combined_df["year"] = combined_df["transit_timestamp"].dt.year
combined_df["month"] = combined_df["transit_timestamp"].dt.month
combined_df["day"] = combined_df["transit_timestamp"].dt.day
combined_df["hour"] = combined_df["transit_timestamp"].dt.hour

# 5. Create a 'weekend_flag' column (1 for weekend, 0 for weekday)
combined_df["weekend_flag"] = combined_df["transit_timestamp"].dt.weekday >= 5  # 0-4 are weekdays, 5-6 are weekend

# Quick preview of the modified DataFrame
print(combined_df.head())

```
