```python
import pandas as pd
import glob
import os

# Set your folder path here
folder_path = "C:/Users/kylea/iCloudDrive/EAE/Data Science/final project/ny_metro_manhattan_chunks/raw_data"

# Get all CSV file paths in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Read and concatenate all CSVs into a single DataFrame
combined_df = pd.concat(
    (pd.read_csv(f) for f in csv_files),
    ignore_index=True
)

# Optional: quick preview
print(f"✅ Loaded {len(csv_files)} CSVs into a DataFrame with {combined_df.shape[0]:,} rows and {combined_df.shape[1]} columns.")
combined_df.head()
```
